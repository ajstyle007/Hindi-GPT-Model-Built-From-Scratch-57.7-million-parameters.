```python
import sentencepiece as spm
```

```python
sp = spm.SentencePieceProcessor()
sp.load("hindi_tokenizer_new.model")
```

```python
words = [
    "संदर्भ",
    "प्रसंग",
    "पृष्ठभूमि",
    "परिप्रेक्ष्य",
    "सापेक्ष",
]

for w in words:
    print(w, "->", sp.encode(w, out_type=str))
```

```python
phrases = [
    "ऊपर दिया गया पाठ",
    "निम्नलिखित अनुच्छेद के आधार पर",
    "उपरोक्त विवरण के अनुसार",
    "पूर्व जानकारी",
    "दी गई जानकारी"
]

for p in phrases:
    print("\n", p)
    print(sp.encode(p, out_type=str))
```

```python
vocab_size = sp.get_piece_size()

for piece in ["▁संदर्भ", "▁प्रसंग", "▁ऊपर", "▁निम्नलिखित"]:
    pid = sp.piece_to_id(piece)
    print(piece, "->", pid if pid != -1 else "NOT IN VOCAB")
```
