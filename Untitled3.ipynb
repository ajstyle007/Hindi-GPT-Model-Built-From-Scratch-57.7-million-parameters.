```python
# Load pretrained model
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
from decoder_only_gpt import My_GPT_model

model = My_GPT_model(vocab_size=sp.get_piece_size(), num_layers=12,
    d_model=512, d_ff=2048, num_heads=8,seq_len=512
).to(DEVICE)
```

```python
import torch
dummy_ids = torch.tensor([[sp.bos_id(), 100, 200, 300]], device=DEVICE)  # small sequence
with torch.no_grad():
    logits = model(dummy_ids)
print("Logits shape:", logits.shape)  # Should be (1, 4, vocab_size)
```

```python
@torch.no_grad()
def simple_generate(prompt_ids, max_new=50):
    input_ids = torch.tensor([prompt_ids], dtype=torch.long).to(DEVICE)
    for _ in range(max_new):
        logits = model(input_ids)
        next_token = torch.argmax(logits[:, -1, :], dim=-1).unsqueeze(0)
        input_ids = torch.cat([input_ids, next_token], dim=1)
        if next_token.item() == sp.eos_id():
            break
    return sp.decode(input_ids[0].tolist())

# Test
test_prompt = sp.encode("महात्मा गांधी ")
print(simple_generate(test_prompt))
```

```python
def build_prompt_ids(context_ids, question):
    instruction = """
आप दिए गए दस्तावेज़ के विशेषज्ञ हैं।
नीचे दिए गए संदर्भ के आधार पर ही उत्तर दीजिए।
⚠️ यदि संदर्भ में जानकारी उपलब्ध नहीं है,
तो केवल यह लिखिए:
"संदर्भ में संबंधित जानकारी उपलब्ध नहीं है।"
किसी भी स्थिति में संदर्भ को दोहराएँ नहीं या अतिरिक्त जानकारी न जोड़ें।
उत्तर हिंदी में ही दें।
"""
    # BOS sirf ek baar shuru mein
    prompt_parts = [
        sp.encode("<s>"),           # agar tera tokenizer <s> ko BOS maanta hai
        sp.encode(instruction.strip()),
        sp.encode("\n\nसंदर्भ:\n"),
        context_ids,
        sp.encode("\n\nप्रश्न:\n"),
        sp.encode(question),
        sp.encode("\n\nउत्तर:\n")
    ]
    
    prompt_ids = []
    for part in prompt_parts:
        prompt_ids.extend(part)
    
    # Agar length > 512 toh last 512 tokens rakh (context sacrifice ho sakta hai)
    if len(prompt_ids) > 512:
        prompt_ids = prompt_ids[-512:]
        
    return prompt_ids



question = "सत्याग्रह क्या है और गांधीजी ने इसे कहाँ से शुरू किया?"

# Chunk
chunks = sentence_chunk_hindi(pdf_text)
    
# Build index
index = build_faiss_index(chunks)

context_ids = retrieve_context(question, chunks, index)

Pythonprompt_ids = build_prompt_ids(context_ids, question)
print("Prompt length:", len(prompt_ids))
print("Decoded prompt start:", sp.decode(prompt_ids[:100]))
```
